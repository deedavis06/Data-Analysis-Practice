```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

# Now for some data visualization! 
In these examples we will be using np.random.randn() to generate random values per normal distribution

First we will start by making 4 different colored line graphs

```python
plots = plt.subplots(nrows=2,ncols=2,figsize=(16,7))
fig, ((ax1, ax2), (ax3,ax4)) = plots
```

```python
ax1.plot(np.random.randn(50), c='blue', linestyle='--', linewidth=3.0)
ax2.plot(np.random.randn(50), c='green', linestyle=':', linewidth=3.0)
ax3.plot(np.random.randn(50), c='yellow', linestyle='-', linewidth=3.0)
ax4.plot(np.random.randn(50), c='red', linestyle='-.', linewidth=3.0)

fig
```

Now lets make some scatterplots!  

```python
fig2 = plt.subplots(nrows=2, ncols=1, figsize=(10,10))
fig, ((ax1,ax2)) = fig2
```

Time to make some scatterplots!

```python
X = np.random.randn(50)
Y = np.random.randn(50) 

ax1.scatter(X,Y, c='yellow')
ax2.scatter(X,Y, c='red')
fig
```

Lets plot a histogram with a randomly generated array of 500 values, value concentration around 350 and a standard deviation of 50.

```python
values= np.random.normal(350,50,500)
fig3,ax=plt.subplots(figsize=(15,8))
fig3=plt.hist(values, bins=50, color='green', ec='black', lw=3)
```

Lets try seaborn! 
Seaborn is a data visualization library built on top of matplotlib that updates the run commands (RC) parameters.
 
Lets make a figure based off the previous variable, 'values' and add a kernal density estimation plot (KDE) to estimate probability distribution.
- *since we set a value of 350 to be value concentration, thats where we should expect to see the peak of the KDE plot*

```python
sns.histplot(data=values, bins=50,color='green', kde=True)
```
